```python
import pandas as pd
import numpy as np
import random 

def gen_uniq_floats(lo, hi, n):
    seen = set()
    needed = n
    while needed != 0:
        uniqs = {random.uniform(lo, hi) for _ in range(needed)}
        seen.update(uniqs)
        needed -= len(uniqs)
    return list(seen)

#l=random.sample(range(10, 100), 50)
#l.sort()

#dtst = {'Time(days)':l}

#df = pd.DataFrame(dtst)
#print(df)
#df.set_index('Time(days)',inplace=True)
#df.to_csv('trainhack.csv')
df = pd.read_csv('trainhack.csv')
#print(df2.head())

#l=df['Time(days)']
l=gen_uniq_floats(0.026, 0.098, 18)
df['Ethylene_concppm'] = l
    
print(df)
        
        
       
```
